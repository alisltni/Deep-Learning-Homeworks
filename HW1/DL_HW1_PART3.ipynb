```python

# -------------------------------------------------------------
# 1. Import required libraries
# -------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# -------------------------------------------------------------
# 2. Load the Iris dataset
# -------------------------------------------------------------
iris = load_iris()
X = iris.data
y = iris.target
print("Shape of dataset:", X.shape)
print("Feature names:", iris.feature_names)
```

```python
# -------------------------------------------------------------
# 3. Data preprocessing (Scaling)
# -------------------------------------------------------------
scaler = StandardScaler()
# TODO: fit and transform X using the scaler
X_scaled = scaler.fit_transform(X)
print("First 5 rows after scaling:\n", X_scaled[:5])
```

```python
# -------------------------------------------------------------
# 4. Determine the optimal number of clusters using Elbow Method
# -------------------------------------------------------------
inertia = []
K_range = range(1, 11)

for k in K_range:
    # TODO: create a KMeans model with k clusters and fit it to X_scaled
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

# Plot the elbow curve
plt.figure(figsize=(7, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

```

```python
#-------------------------------------------------------------
# 5. Apply KMeans with the chosen number of clusters
# -------------------------------------------------------------
# TODO: replace k with the chosen number (e.g., 3)
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# TODO: get predicted cluster labels
y_pred = kmeans.labels_
```
